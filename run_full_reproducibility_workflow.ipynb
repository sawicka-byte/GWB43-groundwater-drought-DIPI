# Full reproducibility workflow

This notebook provides the main entry point for the reproducible analytical workflow used in the study.

It documents:
- repository structure
- required input files
- execution order
- expected outputs
- reproducibility notes

## Repository structure and required inputs

The workflow assumes the following repository structure:

- `data_input/` – input data used for preprocessing
- `data_processed/` – processed analytical outputs
- `figure_source_data/` – source data used for manuscript figures
- `notebooks/` – analytical notebooks executed in sequence

The workflow uses the processed and shareable datasets included in this repository. Raw institutional groundwater monitoring data provided by the Polish Hydrogeological Survey (PIG-PIB) are not redistributed.

```python
from pathlib import Path

# Define repository root assuming this notebook is run from notebooks/
repo_root = Path("..").resolve()

required_files = [
    repo_root / "data_input" / "groundwater_levels_monthly.csv",
    repo_root / "data_input" / "metadata_piezometers.csv",
    repo_root / "data_input" / "precipitation_monthly.csv",
    repo_root / "README.md",
    repo_root / "requirements.txt",
]

print("Checking required files:\n")

all_ok = True
for file in required_files:
    if file.exists():
        print(f"[OK] {file.name}")
    else:
        print(f"[MISSING] {file}")
        all_ok = False

if all_ok:
    print("\nAll required core files are available.")
else:
    print("\nSome required files are missing. Please verify the repository structure and filenames.")
```

## Execution order

The analytical workflow should be run in the following order:

1. `01_preprocessing_SPI_SGI.ipynb`  
   Preparation of monthly groundwater and precipitation series, and calculation of SPI and SGI datasets.

2. `02_event_metrics.ipynb`  
   Derivation of drought event metrics and cluster-level summaries from SGI time series.

3. `03_DIPI_construction.ipynb`  
   Construction of drought vulnerability components and final DIPI-related outputs.

Running the notebooks in this order ensures that all intermediate outputs required by subsequent steps are available.

```python
workflow_notebooks = [
    "01_preprocessing_SPI_SGI.ipynb",
    "02_event_metrics.ipynb",
    "03_DIPI_construction.ipynb",
]

print("Analytical notebooks to run:\n")
for i, nb in enumerate(workflow_notebooks, start=1):
    nb_path = repo_root / "notebooks" / nb
    status = "OK" if nb_path.exists() else "MISSING"
    print(f"{i}. {nb} [{status}]")
```

```python
required_directories = [
    repo_root / "data_input",
    repo_root / "data_processed",
    repo_root / "figure_source_data",
    repo_root / "notebooks",
]

print("Checking required directories:\n")

for directory in required_directories:
    if directory.exists():
        print(f"[OK] {directory.name}/")
    else:
        print(f"[MISSING] {directory}")
```
